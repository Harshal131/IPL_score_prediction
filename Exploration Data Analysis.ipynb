```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv('Dataset/data.csv')
```

```python
col=['runs', 'wickets', 'overs', 'runs_last_5', 'wickets_last_5', 'striker',
       'non-striker']
for i in col:
    plt.title('Distribution of {} column'.format(i))
    df[i].hist()
    plt.show()
```

```python
for i in col:
    plt.title(' Outlier detection of {} '.format(i))
    plt.boxplot(df[i])
    plt.show()
```

```python
sns.pairplot(df,diag_kind='kde',palette="husl")
```

```python
cmap = sns.diverging_palette(220, 10, as_cmap=True)
corr=df.corr()
sns.heatmap(corr,vmax=.3, center=0, cmap=cmap,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```
